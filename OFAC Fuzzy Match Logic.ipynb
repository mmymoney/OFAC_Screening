```python
#dependencies
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
#produce dataframe from Treasury's SDN listing and incorporating the applicable column names
sdn_column_names = ('ent_num', 'SDN_Name', 'SDN_Type', 'Program', 'Title', 'Call_Sign', 'Vess_type', 'Tonnage', 'GRT','Vess_flag','Vess_owner', 'remarks')
sdn_list = pd.read_csv('https://www.treasury.gov/ofac/downloads/sdn.csv', names=sdn_column_names)

non_sdn_column_names = ('ent_num', 'SDN_Name', 'SDN_Type', 'Program', 'Title', 'Call_Sign', 'Vess_type', 'Tonnage', 'GRT','Vess_flag','Vess_owner', 'remarks')
non_sdn_list = pd.read_csv('https://www.treasury.gov/ofac/downloads/consolidated/cons_prim.csv', names = non_sdn_column_names)
```

```python
#read in file for screening
names = pd.read_csv(input('What is the name of your target file? (e.g., MOCK_DATA.csv)'))
names
```

```python
#nested for loop to apply fuzzy match logic against SDN listed names
#and form failed screening list for additional validation
scored_names = []
scored_sdns = []
scored_fuzzies = []
failed_names = []
failed_sdns = []


for name in names[input('What is the column you would like to search the SDN list against (e.g., full_name)?')]:
    for sdn in sdn_list['SDN_Name']:
        try:
            scored_fuzzies.append(fuzz.ratio(name, sdn))
            scored_names.append(name)
            scored_sdns.append(sdn)
#             print(f'There is a {fuzz.ratio(name, sdn)} percent match between the searched name, {name}, and the following SDN name, {sdn}.')
        except TypeError:
            failed_names.append(name)
            failed_sdns.append(sdn)
            print(f'Searched Name: {name}; SDN List Name: {sdn}')
```

```python
#nested for loop to apply fuzzy match logic against SDN listed names
#and form failed screening list for additional validation
scored_non_sdn_names = []
scored_non_sdns = []
scored_non_sdn_fuzzies = []
failed_non_sdn_names = []
failed_non_sdns = []

for name in names[input('What is the column you would like to search the non-SDN list against (e.g., full_name)?')]:
    for non_sdn in non_sdn_list['SDN_Name']:
        try:
            scored_non_sdn_fuzzies.append(fuzz.ratio(name, non_sdn))
            scored_non_sdn_names.append(name)
            scored_non_sdns.append(non_sdn)
#             print(f'There is a {fuzz.ratio(name, sdn)} percent match between the searched name, {name}, and the following SDN name, {sdn}.')
        except TypeError:
            failed_non_sdn_names.append(name)
            failed_non_sdns.append(non_sdn)
            print(f'Searched Name: {name}; non-SDN List Name: {non_sdn}')
```

```python
#form summary dataframes of scored and failed fuzzy match attempts
scored_sdn_df = pd.DataFrame({'Name':scored_names, 'SDN_Name':scored_sdns, 'match_score':scored_fuzzies})
failed_sdn_df = pd.DataFrame({'Failed Name':failed_names, 'Failed SDN_Name':failed_sdns})
len(failed_sdn_df)
```

```python
#form summary dataframes of scored and failed fuzzy match attempts
scored_non_sdn_df = pd.DataFrame({'Name':scored_non_sdn_names, 'SDN_Name':scored_non_sdns, 'match_score':scored_non_sdn_fuzzies})
failed_non_sdn_df = pd.DataFrame({'Failed Name':failed_non_sdn_names, 'Failed SDN_Name':failed_non_sdns})
len(failed_non_sdn_df)
```

```python
#apply filter logic and sort screened names by match score (descending)
near_matches = scored_sdn_df.loc[scored_sdn_df['match_score'] > int(input('What is your target match threshold (0 to 100)?'))]
near_matches = near_matches.sort_values('match_score',ascending=False)
near_matches
```

```python
#apply filter logic and sort screened names by match score (descending)
near_non_sdn_matches = scored_non_sdn_df.loc[scored_non_sdn_df['match_score'] > int(input('What is your target match threshold (0 to 100)?'))]
near_non_sdn_matches = near_non_sdn_matches.sort_values('match_score',ascending=False)
near_non_sdn_matches
```

```python
#define nullchecker function to validate all entries in failed screening were due to NaN values
def nullchecker(df):
    if len(df) == len(df.isnull()):
        print(f'All {len(df)} entries accounted for.')
        
nullchecker(failed_sdn_df)
```
